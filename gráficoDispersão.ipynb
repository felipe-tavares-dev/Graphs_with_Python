```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('trees.csv')
base
```

```python
# Gráfico de dispersão que é utilizado para ver a relação entre duas variáveis
plt.scatter(x=base['Girth'], y=base['Volume'], color='blue', facecolor='red')
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
plt.show()
```

```python
#Gráfico de dispersão sem o marker irá me gerar um gráfico com uma linha
plt.plot(base.Girth, base.Volume)
plt.title('Gráfico de arvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
plt.show()
```

```python
#Gráfico de dispersão com 'afastamento' dos dados (jitter)
#Aqueles dados que estão muito proximos podemos trata-los utilizando o jitter para não deixar o dado em cima do outro
#Fit_reg linha de regressão
sns.regplot(x=base.Girth, y=base.Volume, x_jitter=0.3, fit_reg=True)
```
