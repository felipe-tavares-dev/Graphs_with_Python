```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

#Avaliação dos pesos de cada frango utilizando o histograma


```python
tree = pd.read_csv('trees.csv')
tree
```

```python
#Histograma com 10 divisões (bins) e com gráficos de densidade (kernel density - suavização)

"""
O stat me permite fazer uma analise estatistica para o eixo Y

stat: count, frequency, probability, percent, density
"""

srn.histplot(tree.Volume, color ='blue', kde=True, stat = 'count',  bins=10,).set(title='Volume') 
```

```python
#Carregamento de outra base dados
chicken = pd.read_csv('chicken.csv')
chicken.head()
```

```python
# Criação de um novo dataframe agrupando o atributo 'feed'
agrupado = chicken.groupby(['feed'])['weight'].sum()
agrupado.to_frame()
```

```python
# Novo DataFrame somente para testar os filtros do pandas
teste = chicken.loc[chicken['feed'] == 'horsebean']
teste
```

```python
# Histograma considerado somente o valor 'horsebean'
# Apenas o gráfico de densidade
srn.kdeplot(chicken.loc[chicken['feed'] == 'horsebean']).set(title='horsebean', xlabel='weight')
```

```python
# Histograma considerado somente o valor 'casein'
srn.histplot(chicken.loc[chicken['feed'] == 'casein'], kde=True).set(title='casein', xlabel='weight', ylabel='chickens')
```

```python
# Histograma considerando somente 'soybean'
srn.histplot(chicken.loc[chicken['feed'] == 'soybean'], kde=True).set(title='Soybean', xlabel='weight', ylabel ='Chickens')
```

```python
# Histograma considerando o valor de 'linseed'
srn.histplot(chicken.loc[chicken['feed'] == 'linseed'], kde = True).set(xlabel = 'weight', ylabel= 'Chickens', title = 'linseed')
```

```python
# Histograma considerando o valor do 'feed' como 'meatmeal'
srn.histplot(chicken.loc[chicken['feed'] == 'meatmeal'], kde = True).set(xlabel = 'weight', ylabel= 'Chickens', title = 'meatmeal')
```

```python
#Histograma para considerar o feed como 'sunflower'
srn.histplot(chicken.loc[chicken['feed'] == 'sunflower'], kde = True).set(xlabel = 'weight', ylabel= 'Chickens', title='Sunflower')
```

```python
# Histograma com o feed sendo considerado como 'horsebean'
srn.histplot(chicken.loc[chicken['feed'] == 'horsebean'], kde = True).set(title= 'horsebean', xlabel = 'weight', ylabel='Chickens')
```

```python
plt.figure()
plt.subplot(3,2,1)
srn.histplot(chicken.loc[chicken['feed'] == 'casein'], kde=True).set(title='casein', xlabel='weight', ylabel='chickens')
plt.subplot(3,2,2)
srn.histplot(chicken.loc[chicken['feed'] == 'soybean'], kde=True).set(title='Soybean', xlabel='weight', ylabel ='Chickens')
plt.subplot(3,2,3)
srn.histplot(chicken.loc[chicken['feed'] == 'sunflower'], kde = True).set(xlabel = 'weight', ylabel= 'Chickens', title='Sunflower')
plt.subplot(3,2,4)
srn.histplot(chicken.loc[chicken['feed'] == 'linseed'], kde = True).set(xlabel = 'weight', ylabel= 'Chickens', title = 'linseed')
plt.subplot(3,2,5)
srn.histplot(chicken.loc[chicken['feed'] == 'meatmeal'], kde =True).set(xlabel = 'weight', ylabel= 'Chickens', title = 'meatmeal')
plt.subplot(3,2,6)
srn.histplot(chicken.loc[chicken['feed'] == 'horsebean'], kde = True).set(title= 'horsebean', xlabel = 'weight', ylabel='Chickens')

plt.tight_layout()
```
