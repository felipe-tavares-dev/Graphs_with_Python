```python
import pandas as pd
import seaborn as sns
import statistics as sts
import matplotlib.pyplot as plt
```

```python
tree = pd.read_csv('trees.csv')
tree
```

```python
#vert -> serve para inclionar o boxplot para vertical
# showfilers -> serve remover os outliners

#patch_artist -> serve para colorir o gráfico

plt.boxplot(tree.Volume, vert=False, showfliers=False, notch=True, patch_artist=True)
plt.title("Árvores")
plt.xlabel("Volume")
plt.show()
```

```python
#Plotando todos as colunas disponíveis com o boxplot

plt.boxplot(tree)
plt.title('Árvores')
plt.xlabel('Dados')
```

```python
#Plotando 3 gráficos com cada um deles mostrando informações diferentes
plt.boxplot(tree.Volume, vert = False)
plt.boxplot(tree.Girth, vert=False)
plt.boxplot(tree.Height, vert=False )
plt.title('Arvores')
plt.xlabel('Dados')
plt.show()
```
