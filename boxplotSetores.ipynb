```python
import pandas as pd
```

```python
insect = pd.read_csv('insect.csv')
insect.head()
```

```python
insect.shape
```

```python
#Agrupando os dados baseando no atributo 'spray', contando e somando os registros
agrupado = insect.groupby(['spray'])['count'].sum()
agrupado
```

```python
#Graficos de barras
agrupado.plot.bar(color = 'skyblue')
```

```python
#Cores
agrupado.plot.bar(color = ['blue', 'yellow', 'red', 'green', 'pink', 'orange'])
```

```python
#Gr√°ficos de pizza
agrupado.plot.pie()
```

```python
#Com legenda
agrupado.plot.pie(legend = True)
```
