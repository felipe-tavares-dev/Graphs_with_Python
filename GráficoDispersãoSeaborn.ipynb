```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
co2 = pd.read_csv('co2.csv')
co2
```

```python
#Gráficos de dispersão utilizando os atributos conc e uptake, agrupado pelo type
#Hue: Irá agrupar as variáveis da coluna Type e irá colocar uma cor para cada variável
sns.scatterplot(data = co2, x='conc', y = 'uptake', hue='Type')
```

```python
#Selecão de registros especificas dos dados base (Quebec e Missipi)
q = co2.loc[co2['Type']=='Quebec']
m = co2.loc[co2['Type'] == 'Mississippi']
```

```python
#Subgrafico (Uma linha e duas colunas) mostrando gráficos sobre a região
uniques = list(co2['Type'].unique())
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(x=q.conc,y=q.uptake).set(title=f'{uniques[0]}')
plt.subplot(1,2,2)
sns.scatterplot(x=m.conc,y=m.uptake).set(title=f'{uniques[1]}')
plt.tight_layout()
```

```python
#Resfriado e não resfriado
ch = co2.loc[co2['Treatment'] == 'chilled']
nc = co2.loc[co2['Treatment'] == 'nonchilled']
```

```python
# Gráficos somente com 'chilled' e 'non-chilled'
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(x=ch.conc, y=ch.uptake).set(title = f'Chilled')
plt.subplot(1,2,2)
sns.scatterplot(x=nc.conc, y=nc.uptake).set(title=f'non-chilled')
plt.tight_layout()
```
