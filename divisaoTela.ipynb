```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tree = pd.read_csv('trees.csv')
tree
```

```python
#Girth com volume 
plt.scatter(tree.Girth, tree.Volume)
```

```python
#Gerando um gráfico de dispersão com o seaborn do Volume e circunferência
sns.regplot(x=tree.Girth, y=tree.Volume, fit_reg=True, x_jitter=0.3)
```

```python
#Gerando um gráfico de dispersão com plotly
plt.scatter(x=tree.Girth, y=tree.Height)
```

```python
#Girth com height usando seaborn para ver a regressão
sns.regplot(x=tree.Girth, y=tree.Height, data = tree, fit_reg=True)
```

```python
#Gráfico de dispersão com altura e volume
plt.scatter(tree.Height, tree.Volume)
```

```python
sns.regplot(x=tree.Height, y = tree.Volume, fit_reg=True, x_jitter=0.3)
```

```python
plt.hist(tree.Volume)
plt.show()
```

```python
#Imprimindo todos os gráficos juntos para podermos ter uma melhor visualização
#Parâmetro figure serve para criarmos uma figura, onde desejamos colocar uma figura

# O subplot permite criarmos uma matriz que irá dividir as nossas imagens

# O primeiro parêmetro que colocamos é o número de linhas 

# plt.subplot(nrows)

# O segundo o número de colunas e o terceiro a posição do gráfico(index)
# plt.subplot(nrows, ncol, position)

plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(tree.Girth, tree.Volume)
plt.subplot(2,2,2)
plt.scatter(tree.Girth, tree.Height)
plt.subplot(2,2,3)
plt.scatter(tree.Height, tree.Volume)
plt.subplot(2,2,4)
plt.hist(tree.Volume)
plt.show()
```
